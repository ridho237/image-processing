#3.Creation of Image watermark

```python
# Ukuran watermark
size = (350, 350)

# Load gambar utama
image = Image.open("/content/zzz.jpg").convert("RGBA")

# Load watermark dan konversi ke RGBA
watermark = Image.open("/content/miyabi.jpg").convert("RGBA")
watermark.thumbnail(size)

# Buat transparansi pada watermark
alpha = 0.7  # Nilai alpha (0 = transparan penuh, 1 = tidak transparan)
watermark_with_alpha = Image.new("RGBA", watermark.size)
for x in range(watermark.width):
    for y in range(watermark.height):
        r, g, b, a = watermark.getpixel((x, y))
        watermark_with_alpha.putpixel((x, y), (r, g, b, int(a * alpha)))

# Salin gambar utama dan tambahkan watermark
copied_image = image.copy()
copied_image.paste(watermark_with_alpha, (75, 280), watermark_with_alpha)

# Tampilkan hasil
plt.axis(False)
plt.imshow(copied_image)
plt.show()
```
